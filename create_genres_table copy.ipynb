```python
import pandas as pd
import numpy as np
```

```python
movies_df = pd.read_csv("raw_data/tmdb_movies.csv")
movies_df.head()
```

```python
movies_df = movies_df.drop_duplicates()
movies_df.nunique()
```

```python
genres_df = movies_df[["id","genres"]]
genres_df.head()
```

```python
import ast
```

```python
def convert(text):
    L = []
    for i in ast.literal_eval(text):
        L.append((i['id'], i['name']))
    return L
```

```python
genres_df2 = genres_df
genres_df['genres'] = genres_df['genres'].apply(convert)
genres_df.head()
```

```python
import re
delimiter = ','
genres_df['genres'] = genres_df['genres'].astype(str)

print(genres_df)
```

```python
genres_df2 = genres_df['genres'].str.split(delimiter, expand=True)
```

```python
id_column = genres_df['id']
genres_df2 = pd.concat([id_column, genres_df2], axis=1)
genres_df2.head()
```

```python
genres_df2.to_csv("clean_data/genres.csv", index=False)
```

```python
genres_df3 = pd.read_csv("clean_data/genres.csv")
genres_df3.head()
```

```python
genres_df3 = genres_df3.rename(columns={str(i): f'col_{i}' for i in range(14)})
```

```python
chars_to_strip = " '123._/*()[]'"
for i in range(14):  # 14 because we want 'col_0' to 'col_13'
    col_name = f"col_{i}"
    genres_df3[col_name] = genres_df3[col_name].str.strip(chars_to_strip)
```

```python
genres_df3.head()
```

```python
# Checked all columns for unique values, every pair column (col_2, col_4, col_6, etc...) has exclusively genre_id;
# To drop all the columns containing the genre_id (unnecessary data):
columns_to_drop = [f'col_{i}' for i in range(0, 14, 2)]

genres_df3.drop(columns_to_drop, axis=1, inplace=True)
```

```python
renaming_dict = {f'col_{i}': f'genre_{(i // 2) + 1}' for i in range(1, 14, 2)}

genres_df3 = genres_df3.rename(columns=renaming_dict) # Renaming the columns using the generated dictionary
```

```python
genres_df3.head()
```

```python
genres_columns = [f'genre_{i}' for i in range(1, 8)]

genres_df3.update(genres_df3[genres_columns].fillna(" ")) # Fill NaN values with an empty space for these columns
```
